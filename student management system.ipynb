### STUDENT MANAGEMENT SYSTEM

```python
# -----------------------------------------------Importing Libraries-----------------------------------------------------------
import mysql.connector
from mysql.connector import Error
import tkinter
from tkinter import *
from tkinter import ttk
from tkinter import messagebox as mb
from PIL import ImageTk

# ---------------------------------------------Creating Tkinter Window---------------------------------------------------------        
root=Tk()
root.title("STUDENT DATABASE MANAGEMENT SYSTEM")
root.attributes('-fullscreen', True)

# ------------------------------------------------Creating Treeview------------------------------------------------------------
treeview = ttk.Treeview(root, columns=('ROLL NO.', 'NAME', "FATHER'S NAME", 'DOB', 'CONTACT', 'ADDRESS'))
treeview_width = root.winfo_screenwidth() - 350
treeview.place(x=350,y=140,width=treeview_width,height=723)

# ---------------------------------------------Treeview column widths----------------------------------------------------------
treeview.column("#0", width=0, stretch=NO) 
treeview.column("ROLL NO.", width=int(treeview_width * 0.1))
treeview.column("NAME", width=int(treeview_width * 0.15))
treeview.column("FATHER'S NAME", width=int(treeview_width * 0.15))
treeview.column("DOB", width=int(treeview_width * 0.1))
treeview.column("CONTACT", width=int(treeview_width * 0.15))
treeview.column("ADDRESS", width=int(treeview_width * 0.35))

# ---------------------------------------------Treeview column headings---------------------------------------------------------
treeview.heading("#0", text="", anchor=W)
treeview.heading("ROLL NO.", text="ROLL NO.", anchor="center")
treeview.heading("NAME", text="NAME", anchor="center")
treeview.heading("FATHER'S NAME", text="FATHER'S NAME", anchor="center")
treeview.heading("DOB", text="DOB", anchor="center")
treeview.heading("CONTACT", text="CONTACT", anchor="center")
treeview.heading("ADDRESS", text="ADDRESS", anchor="center")

# ----------------------------------------------Treeview Scrollbar-------------------------------------------------------------
scrollbar = ttk.Scrollbar(root, orient="vertical", command=treeview.yview)
scrollbar.place(x=treeview_width + 350 - 20, y=140, height=723)
treeview.configure(yscrollcommand=scrollbar.set)

# ------------------------------------------------Creating Frames-------------------------------------------------------------
topframe=Frame(root,bg='black',width=1550,height=100)
topframe.place(x=0,y=0)
leftframe=Frame(root,bg='black',width=350,height=790)
leftframe.place(x=0,y=100)
top2frame=Frame(root,bg='light grey',width=1550,height=40)
top2frame.place(x=350,y=100)

# ------------------------------------------------Creating Lables--------------------------------------------------------------

mainlabel=Label(root,text="STUDENT MANAGEMENT SYSTEM",font=('Times New Roman',50,'bold','underline'),bg='black',fg='cyan')
mainlabel.place(x=200,y=5)


l1=Label(root,text='ROLL NO',font=('Arial',15,'bold'),width=8,bg='black',fg='cyan')
l1.place(x=0,y=180)
e1=Entry(root,width=17,bg='white',fg='black',font=('bold'))
e1.place(x=150,y=185,height=30)

l2=Label(root,text='NAME',font=('Arial',15,'bold'),width=5,bg='black',fg='cyan')
l2.place(x=0,y=250)
e2=Entry(root,width=17,bg='white',fg='black',font=('bold'))
e2.place(x=150,y=255,height=30)

l3=Label(root,text="FATHER NAME",font=('Arial',15,'bold'),width=12,bg='black',fg='cyan')
l3.place(x=0,y=320)
e3=Entry(root,width=17,bg='white',fg='black',font=('bold'))
e3.place(x=150,y=325,height=30)

l4=Label(root,text='DOB',font=('Arial',15,'bold'),width=4,bg='black',fg='cyan')
l4.place(x=0,y=390)
e4=Entry(root,width=17,bg='white',fg='black',font=('bold'))
e4.place(x=150,y=395,height=30)

l5=Label(root,text='CONTACT',font=('Arial',15,'bold'),width=8,bg='black',fg='cyan')
l5.place(x=0,y=460)
e5=Entry(root,width=17,bg='white',fg='black',font=('bold'))
e5.place(x=150,y=465,height=30)

l6=Label(root,text='ADDRESS',font=('Arial',15,'bold'),width=8,bg='black',fg='cyan')
l6.place(x=0,y=530)
e6=Entry(root,width=17,bg='white',fg='black',font=('bold'))
e6.place(x=150,y=535,height=40)
l7=Label(root,text="STUDENT INFO",font=('Times New Roman',25,'bold','underline'),bg='black',fg='cyan')
l7.place(x=50,y=110)


# -------------------------------------------------Creating Combobox------------------------------------------------------------

search_options = ["ROLL_NO", "NAME", "CONTACT"]
selected_search_option = StringVar()
search_combobox = ttk.Combobox(root, textvariable=selected_search_option, values=search_options)
search_combobox.place(x=360, y=105, height=27)
search_combobox.current(0)

# ------------------------------------------------Combobox Entry Box------------------------------------------------------------
searchentry=Entry(root,font=('bold'),width=40)
searchentry.place(x=550,y=105)

#--------------------------------------------------Creating Functions-----------------------------------------------------------

def connection():
    def connectivity(hostname,username,upassword,db_name):
        conn = None
        try:
            conn = mysql.connector.connect(
                host=hostname,
                user=username,
                password=upassword,
                db = db_name
            )
            mb.showinfo("Success","Connection Successfull")
            return conn
        except Error as e:
            mb.showerror("Error",e)
    connectivity('localhost','root','Vishujain@1','student_management_system')
            
def create_tb():
    conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db = 'student_management_system'
            )
    result = conn.cursor()
    try:
        result.execute("""create table if not exists student_rec (
        ROLL_NO varchar(5) primary key not null,
        NAME varchar(100) not null,
        FATHER_NAME varchar(100),
        DOB varchar(30),
        CONTACT varchar(20),
        ADDRESS varchar(100));""")
    except Error as e:
        mb.showerror('Error',e)
        
def add():
    a = e1.get()
    b = e2.get()
    c = e3.get()
    d = e4.get()
    e = e5.get()
    f = e6.get()

    if a == '' or b == '' or c == '' or d == '' or e == '' or f == '':
        mb.showerror("Error", "Fields can't be empty")
    else:
        try:
            conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db='student_management_system'
            )
            cursor = conn.cursor()
            sql = "INSERT INTO student_rec (ROLL_NO, NAME, FATHER_NAME, DOB, CONTACT, ADDRESS) VALUES (%s, %s, %s, %s, %s, %s)"
            cursor.execute(sql, (a, b, c, d, e, f))
            conn.commit()
            mb.showinfo("Success", "Data Added")
            e1.delete(0,END)
            e2.delete(0,END)
            e3.delete(0,END)
            e4.delete(0,END)
            e5.delete(0,END)
            e6.delete(0,END)
        except Error as e:
            mb.showerror('Error',e)
        finally:
            cursor.close()
            conn.close()
            
def update():
    a = e1.get()
    b = e2.get()
    c = e3.get()
    d = e4.get()
    e = e5.get()
    f = e6.get()
    
    if a=='':
        mb.showerror('Error',"Please enter Roll No")
    else:
        try:
            conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db='student_management_system'
            )
            cursor = conn.cursor()
            sql="UPDATE student_rec SET NAME = %s , FATHER_NAME = %s , DOB = %s , CONTACT = %s , ADDRESS = %s where ROLL_NO = %s"
            cursor.execute(sql, (b, c, d, e, f, a))
            conn.commit()
            mb.showinfo("Success", "Data Updated")
            e1.delete(0,END)
            e2.delete(0,END)
            e3.delete(0,END)
            e4.delete(0,END)
            e5.delete(0,END)
            e6.delete(0,END)
        except Error as e:
            mb.showerror('Error',e)
        finally:
            cursor.close()
            conn.close()
            
def delete():
    a = e1.get()
    e1.delete(0,END)
    if a=='':
        mb.showerror('Error',"Please enter Roll No")
    else:
        try:
            conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db='student_management_system'
            )
            cursor = conn.cursor()
            sql="DELETE FROM student_rec where ROLL_NO = %s"
            cursor.execute(sql, (a,))
            conn.commit()
            mb.showinfo("Success", "Data Deleted")
        except Error as e:
            mb.showerror('Error',e)
        finally:
            cursor.close()
            conn.close()
                    
def showdata():
    try:
        conn = mysql.connector.connect(
            host='localhost',
            user='root',
            password='Vishujain@1',
            db='student_management_system'
        )
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM student_rec")
        rows = cursor.fetchall()

        # Clear any existing data in the treeview
        for row in treeview.get_children():
            treeview.delete(row)

        # Insert fetched data into the treeview
        for row in rows:
            treeview.insert('', 'end', values=row)

        conn.close()
    except Error as e:
        print('Error:', e)
        
def searchstud():
    a = selected_search_option.get()  # Get selected search option from combobox
    b = searchentry.get()

    if a == '':
        mb.showerror('Error', 'Please select a search option')
    elif b == '':
        mb.showerror('Error', 'Please enter search detail')
    else:
        try:
            conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db='student_management_system'
            )
            cursor = conn.cursor()
            sql = "SELECT * FROM student_rec WHERE `{}` = %s".format(a)  # Enclose column name in backticks
            cursor.execute(sql, (b,))
            rows = cursor.fetchall()
            for row in treeview.get_children():
                treeview.delete(row)
            for row in rows:
                treeview.insert('', 'end', values=row)
        except Error as e:
            mb.showerror('Error', e)
        finally:
            cursor.close()
            conn.close()

def clearall():
    e1.delete(0,END)
    e2.delete(0,END)
    e3.delete(0,END)
    e4.delete(0,END)
    e5.delete(0,END)
    e6.delete(0,END)
    for row in treeview.get_children():
        treeview.delete(row)
    
def minimize():
    root.iconify()    
        
def exit():
    confirmation= mb.askokcancel("Alert","Arey you sure you want to exit ?")
    if confirmation:
        root.destroy()



#-------------------------------------------------Creating Buttons--------------------------------------------------------------

b1=Button(root,text='ADD',font=('Times New Roman',15,'bold'),width=28,height=2,bg='pink',fg='black',command=add)
b1.place(x=3,y=590)
b2=Button(root,text='UPDATE',font=('Times New Roman',15,'bold'),width=28,height=2,bg='pink',fg='black',command=update)
b2.place(x=3,y=660)
b3=Button(root,text='DELETE',font=('Times New Roman',15,'bold'),width=28,height=2,bg='pink',fg='black',command=delete)
b3.place(x=3,y=730)
b4=Button(root,text='CLEAR ALL',font=('Times New Roman',15,'bold'),width=28,height=2,bg='pink',fg='black',command=clearall)
b4.place(x=3,y=800)
b5=Button(root,text='Connect',font=('Times New Roman',10,'bold'),bg='white',fg='red',width=10,height=2,command=lambda: [connection(),create_tb()])
b5.place(x=1385,y=0)
b6=Button(root,text='SEARCH',font=('Times New Roman',10,'bold'),width=15,height=1,bg='light grey',fg='black',command=searchstud)
b6.place(x=1050,y=105,height=27)
b7=Button(root,text='SHOW ALL',font=('Times New Roman',10,'bold'),width=15,height=1,bg='light grey',fg='black',command=showdata)
b7.place(x=1200,y=105)
b8=Button(root,text='X',font=('bold'),command=exit,bg='white',fg='red',width=2)
b8.place(x=1505,y=0)
b9=Button(root,text='-',font=('bold'),command=minimize,bg='white',fg='red',width=2)
b9.place(x=1470,y=0)

#-------------------------------------------------END OF THE CODE--------------------------------------------------------------

root.mainloop()
```
