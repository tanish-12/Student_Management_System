```python
                                                    ### MAIN CODE ### 

import mysql.connector
from mysql.connector import Error
import tkinter
from tkinter import *
from tkinter import ttk
from tkinter import messagebox as mb
from PIL import ImageTk
        
root=Tk()
root.title("STUDENT DATABASE MANAGEMENT SYSTEM")
root.minsize(1400,730)
root.resizable('False','False')
column= ('Roll No', 'Name', 'Father Name', 'DOB', 'Contact', 'Address')
treeview = ttk.Treeview(root)
treeview.place(x=350,y=50)
        
f1=Frame(root,bg='red',width=350,height=680)
f1.place(x=0,y=50)
f2=Frame(root,bg='red',width=350,height=50)
f2.place(x=0,y=0)
f3=Frame(root,bg='red',width=1050,height=50)
f3.place(x=350,y=0)
#f4=Frame(root,bg='pink',width=1050,height=680)
#f4.place(x=350,y=50)
detail=Label(root,text='STUDENT DETAILS',font=('Times New Roman',18,'underline','bold'),bg='red',fg='black')
detail.place(x=60,y=10)
info=Label(root,text='INFORMATION',font=('Times New Roman',18,'underline','bold'),bg='pink',fg='black')
info.place(x=800,y=10)

#details_labels_entrybox

l1=Label(root,text='ROLL NO',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
l1.place(x=10,y=70)
e1=Entry(root,width=30,bg='white',fg='black')
e1.place(x=150,y=70)
l2=Label(root,text='NAME',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
l2.place(x=10,y=120)
e2=Entry(root,width=30,bg='white',fg='black')
e2.place(x=150,y=120)
l3=Label(root,text="FATHER'S NAME",font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
l3.place(x=10,y=170)
e3=Entry(root,width=30,bg='white',fg='black')
e3.place(x=150,y=170)
l4=Label(root,text='DOB',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
l4.place(x=10,y=220)
e4=Entry(root,width=30,bg='white',fg='black')
e4.place(x=150,y=220)
l5=Label(root,text='CONTACT',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
l5.place(x=10,y=270)
e5=Entry(root,width=30,bg='white',fg='black')
e5.place(x=150,y=270)
l6=Label(root,text='ADDRESS',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
l6.place(x=10,y=320)
e6=Entry(root,width=30,bg='white',fg='black')
e6.place(x=150,y=320)

def connection():
    def connectivity(hostname,username,upassword,db_name):
        conn = None
        try:
            conn = mysql.connector.connect(
                host=hostname,
                user=username,
                password=upassword,
                db = db_name
            )
            mb.showinfo("Success","Connection Successfull")
            return conn
        except Error as e:
            mb.showerror("Error",e)
    connectivity('localhost','root','Vishujain@1','student_management_system')
            
def create_tb():
    conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db = 'student_management_system'
            )
    result = conn.cursor()
    try:
        result.execute("""create table if not exists student_rec (
        Student_Roll_No int(5) primary key not null,
        Student_Name varchar(100) not null,
        Student_Father_Name varchar(100),
        Student_DOB date,
        Student_Contact int(20),
        Student_Address varchar(100));""")
    except Error as e:
        mb.showerror('Error',e)
        
def add():
    a = e1.get()
    b = e2.get()
    c = e3.get()
    d = e4.get()
    e = e5.get()
    f = e6.get()

    if a == '' or b == '' or c == '' or d == '' or e == '' or f == '':
        mb.showerror("Error", "Fields can't be empty")
    else:
        try:
            a=int(a)
            e=int(e)
            conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db='student_management_system'
            )
            cursor = conn.cursor()
            sql = "INSERT INTO student_rec (Student_Roll_No, Student_Name, Student_Father_Name, Student_DOB, Student_Contact, Student_Address) VALUES (%s, %s, %s, %s, %s, %s)"
            cursor.execute(sql, (a, b, c, d, e, f))
            conn.commit()
            mb.showinfo("Success", "Data Added")
            e1.delete(0,END)
            e2.delete(0,END)
            e3.delete(0,END)
            e4.delete(0,END)
            e5.delete(0,END)
            e6.delete(0,END)
        except Error as e:
            mb.showerror('Error',e)
        finally:
            cursor.close()
            conn.close()
            
def update():
    a = e1.get()
    b = e2.get()
    c = e3.get()
    d = e4.get()
    e = e5.get()
    f = e6.get()
    
    if a=='':
        mb.showerror('Error',"Please enter Roll No")
    else:
        try:
            a=int(a)
            e=int(e)
            conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db='student_management_system'
            )
            cursor = conn.cursor()
            sql="UPDATE student_rec SET Student_Name = %s , Student_Father_Name = %s , Student_DOB = %s , Student_Contact = %s , Student_Address = %s where Student_Roll_No = %s"
            cursor.execute(sql, (b, c, d, e, f, a))
            conn.commit()
            mb.showinfo("Success", "Data Updated")
            e1.delete(0,END)
            e2.delete(0,END)
            e3.delete(0,END)
            e4.delete(0,END)
            e5.delete(0,END)
            e6.delete(0,END)
        except Error as e:
            mb.showerror('Error',e)
        finally:
            cursor.close()
            conn.close()
            
def delete():
    a = e1.get()
    e1.delete(0,END)
    if a=='':
        mb.showerror('Error',"Please enter Roll No")
    else:
        try:
            conn = mysql.connector.connect(
                host='localhost',
                user='root',
                password='Vishujain@1',
                db='student_management_system'
            )
            cursor = conn.cursor()
            sql="DELETE FROM student_rec where Student_Roll_No = %s"
            cursor.execute(sql, (a,))
            conn.commit()
            mb.showinfo("Success", "Data Deleted")
        except Error as e:
            mb.showerror('Error',e)
        finally:
            cursor.close()
            conn.close()
                    
def showdata():
    try:
        conn = mysql.connector.connect(
            host='localhost',
            user='root',
            password='Vishujain@1',
            db='student_management_system'
        )
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM student_rec")
        rows = cursor.fetchall()

        # Clear any existing data in the treeview
        for row in tree.get_children():
            tree.delete(row)

        # Insert fetched data into the treeview
        for row in rows:
            tree.insert('', 'end', values=row)

        conn.close()
    except Error as e:
        print('Error:', e)


#buttons
b1=Button(root,text='ADD',font=('Times New Roman',15,'bold'),width=15,height=1,bg='pink',fg='black',command=add)
b1.place(x=80,y=370)
b2=Button(root,text='UPDATE',font=('Times New Roman',15,'bold'),width=15,height=1,bg='pink',fg='black',command=update)
b2.place(x=80,y=420)
b3=Button(root,text='DELETE',font=('Times New Roman',15,'bold'),width=15,height=1,bg='pink',fg='black',command=delete)
b3.place(x=80,y=470)
b4=Button(root,text='Connect to Database',font=('Times New Roman',12,'bold'),bg='pink',command=lambda: [connection(),create_tb()])
b4.place(x=1200,y=8)
b5=Button(root,text='SHOW DATA',font=('Times New Roman',15,'bold'),width=15,height=1,bg='pink',fg='black',command=showdata)
b5.place(x=80,y=520)

root.mainloop()
```
