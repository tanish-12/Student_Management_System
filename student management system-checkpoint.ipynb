```python
                                                    ### MAIN CODE ### 

import mysql.connector
from mysql.connector import Error
import tkinter
from tkinter import *
from tkinter import ttk
from tkinter import messagebox as mb
from PIL import ImageTk


###creating login dialog box
loginroot=Tk()
loginroot.geometry("400x250")
loginroot.resizable('False','False')
loginroot.title("LOGIN")

#background=ImageTk.PhotoImage(file=r'C:\Users\jaint\Desktop\sdbms project\bg.jpg')
#bglabel=Label(loginroot,image=background)
#bglabel.place(x=0,y=0)

def login():
    u=ex1.get()
    p=ex2.get()
    
    if u=='' or p=='':
        mb.showerror("Error","Fields can't be empty !")
        
    elif u=='Tanish' or u=='tanish' and p=='1234':
        mb.showinfo('Success','User logged in succesfully !')
        loginroot.destroy()
        
        root=Tk()
        root.title("STUDENT DATABASE MANAGEMENT SYSTEM")
        root.minsize(1400,730)
        root.resizable('False','False')
        
        tree = ttk.Treeview(root, columns=("Roll No", "Name", "Age", "Grade", "Marks", "Address"), show="headings")
        tree.heading("Roll No", text="Roll No")
        tree.heading("Name", text="Name")
        tree.heading("Age", text="Age")
        tree.heading("Grade", text="Grade")
        tree.heading("Marks", text="Marks")
        tree.heading("Address", text="Address")

        # Pack the treeview widget
        tree.place(x=400,y=200)


        f1=Frame(root,bg='cyan',width=350,height=680)
        f1.place(x=0,y=50)
        f2=Frame(root,bg='pink',width=350,height=50)
        f2.place(x=0,y=0)
        f3=Frame(root,bg='cyan',width=1050,height=50)
        f3.place(x=350,y=0)
        f4=Frame(root,bg='pink',width=1050,height=680)
        f4.place(x=350,y=50)
        detail=Label(root,text='STUDENT DETAILS',font=('Times New Roman',18,'bold','underline'),bg='pink',fg='black')
        detail.place(x=60,y=10)
        info=Label(root,text='INFORMATION',font=('Times New Roman',18,'bold','underline'),bg='cyan',fg='black')
        info.place(x=800,y=10)

        #details_labels_entrybox

        l1=Label(root,text='ROLL NO',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
        l1.place(x=10,y=70)
        e1=Entry(root,width=30,bg='white',fg='black')
        e1.place(x=150,y=70)
        l2=Label(root,text='NAME',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
        l2.place(x=10,y=120)
        e2=Entry(root,width=30,bg='white',fg='black')
        e2.place(x=150,y=120)
        l3=Label(root,text="FATHER'S NAME",font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
        l3.place(x=10,y=170)
        e3=Entry(root,width=30,bg='white',fg='black')
        e3.place(x=150,y=170)
        l4=Label(root,text='DOB',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
        l4.place(x=10,y=220)
        e4=Entry(root,width=30,bg='white',fg='black')
        e4.place(x=150,y=220)
        l5=Label(root,text='CONTACT',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
        l5.place(x=10,y=270)
        e5=Entry(root,width=30,bg='white',fg='black')
        e5.place(x=150,y=270)
        l6=Label(root,text='ADDRESS',font=('Times New Roman',10,'bold'),width=15,bg='white',fg='black')
        l6.place(x=10,y=320)
        e6=Entry(root,width=30,bg='white',fg='black')
        e6.place(x=150,y=320)

        def connection():
            def connectivity(hostname,username,upassword,db_name):
                conn = None
                try:
                    conn = mysql.connector.connect(
                        host=hostname,
                        user=username,
                        password=upassword,
                        db = db_name
                    )
                    mb.showinfo("Success","Connection Successfull")
                    return conn
                except Error as e:
                    print(f"Error : {e}")
            connectivity('localhost','root','Vishujain@1','student_management_system')
            
        def create_tb():
            conn = mysql.connector.connect(
                        host='localhost',
                        user='root',
                        password='Vishujain@1',
                        db = 'student_management_system'
                    )
            result = conn.cursor()
            try:
                result.execute("""create table if not exists student_rec (
                Student_Roll_No int primary key not null,
                Student_Name varchar(30) not null,
                Student_Father_Name varchar(30),
                Student_DOB date,
                Student_Contact int(10),
                Student_Address varchar(50));""")
            except Error as e:
                mb.showerror('Error','Error in Table Creation')
        
        def add():
            a = int(e1.get())
            b = e2.get()
            c = e3.get()
            d = e4.get()
            e = int(e5.get())
            f = e6.get()
            e1.delete(0,END)
            e2.delete(0,END)
            e3.delete(0,END)
            e4.delete(0,END)
            e5.delete(0,END)
            e6.delete(0,END)

            if a == '' or b == '' or c == '' or d == '' or e == '' or f == '':
                mb.showerror("Error", "Fields can't be empty")
            else:
                try:
                    conn = mysql.connector.connect(
                        host='localhost',
                        user='root',
                        password='Vishujain@1',
                        db='student_management_system'
                    )
                    cursor = conn.cursor()
                    sql = "INSERT INTO student_rec (Student_Roll_No, Student_Name, Student_Father_Name, Student_DOB, Student_Contact, Student_Address) VALUES (%s, %s, %s, %s, %s, %s)"
                    cursor.execute(sql, (a, b, c, d, e, f))
                    conn.commit()
                    mb.showinfo("Success", "Data Added")
                except Error as e:
                    print('Error',e)
                    mb.showerror('Error', 'Unable to add data !')
                finally:
                    cursor.close()
                    conn.close()
            
        def update():
            a = int(e1.get())
            b = e2.get()
            c = e3.get()
            d = e4.get()
            e = int(e5.get())
            f = e6.get()
            e1.delete(0,END)
            e2.delete(0,END)
            e3.delete(0,END)
            e4.delete(0,END)
            e5.delete(0,END)
            e6.delete(0,END)
    
            if a=='':
                mb.showerror('Error',"Please enter Roll No")
            else:
                try:
                    conn = mysql.connector.connect(
                        host='localhost',
                        user='root',
                        password='Vishujain@1',
                        db='student_management_system'
                    )
                    cursor = conn.cursor()
                    sql="UPDATE student_rec SET Student_Name = %s , Student_Father_Name = %s , Student_DOB = %s , Student_Contact = %s , Student_Address = %s where Student_Roll_No = %s"
                    cursor.execute(sql, (b, c, d, e, f, a))
                    conn.commit()
                    mb.showinfo("Success", "Data Updated")
                except Error as e:
                    print('Error',e)
                    mb.showerror('Error', 'Unable to update data !')
                finally:
                    cursor.close()
                    conn.close()
            
        def delete():
            a = e1.get()
            e1.delete(0,END)
            if a=='':
                mb.showerror('Error',"Please enter Roll No")
            else:
                try:
                    conn = mysql.connector.connect(
                        host='localhost',
                        user='root',
                        password='Vishujain@1',
                        db='student_management_system'
                    )
                    cursor = conn.cursor()
                    sql="DELETE FROM student_rec where Student_Roll_No = %s"
                    cursor.execute(sql, (a,))
                    conn.commit()
                    mb.showinfo("Success", "Data Deleted")
                except Error as e:
                    print('Error',e)
                    mb.showerror('Error', 'Unable to delete data !')
                finally:
                    cursor.close()
                    conn.close()
                    
        def showdata():
            try:
                conn = mysql.connector.connect(
                    host='localhost',
                    user='root',
                    password='Vishujain@1',
                    db='student_management_system'
                )
                cursor = conn.cursor()
                cursor.execute("SELECT * FROM student_rec")
                rows = cursor.fetchall()

                # Clear any existing data in the treeview
                for row in tree.get_children():
                    tree.delete(row)

                # Insert fetched data into the treeview
                for row in rows:
                    tree.insert('', 'end', values=row)

                conn.close()
            except Error as e:
                print('Error:', e)

    
    
    
    

        #buttons
        b1=Button(root,text='ADD',font=('Times New Roman',15,'bold'),width=15,height=1,bg='pink',fg='black',command=add)
        b1.place(x=80,y=370)
        b2=Button(root,text='UPDATE',font=('Times New Roman',15,'bold'),width=15,height=1,bg='pink',fg='black',command=update)
        b2.place(x=80,y=420)
        b3=Button(root,text='DELETE',font=('Times New Roman',15,'bold'),width=15,height=1,bg='pink',fg='black',command=delete)
        b3.place(x=80,y=470)
        b4=Button(root,text='Connect to Database',font=('Times New Roman',12,'bold'),bg='pink',command=lambda: [connection(),create_tb()])
        b4.place(x=1200,y=8)
        b5=Button(root,text='SHOW DATA',font=('Times New Roman',15,'bold'),width=15,height=1,bg='pink',fg='black',command=showdata)
        b5.place(x=80,y=520)



        root.mainloop()
    else:
        mb.showerror("Error","Please enter valid credentials")
    

user=StringVar()
passw=StringVar()

lx1=Label(loginroot,text="USERNAME :",font=("Times New Roman",12,"bold","underline"),bg='black',fg='yellow')
lx1.place(x=30,y=60)
ex1=Entry(loginroot,width=20,font=("Times New Roman",18,"bold"),textvariable=user,bg='black',fg='yellow')
ex1.place(x=140,y=60)

lx2=Label(loginroot,text="PASSWORD :",font=("Times New Roman",12,"bold","underline"),bg='black',fg='yellow')
lx2.place(x=30,y=100)
ex2=Entry(loginroot,width=20,font=("Times New Roman",18,"bold"),textvariable=passw,bg='black',fg='yellow')
ex2.place(x=140,y=100)

bx1=Button(loginroot,text="LOGIN",font=("Times New Roman",12,"bold"),bg='black',fg='yellow',command=login)
bx1.place(x=170,y=150)


loginroot.mainloop()
```
